```python
import pandas as pd

customers = pd.read_csv (r'/Users/brigettethompson/Desktop/BIP-PYTHON/sales_analysis/data/customers.csv')
print (customers)
```

```python
#change id to customer_id
customers = customers.rename(columns={'id': 'customer_id'})
print(customers)
```

```python
customers.set_index('customer_id', inplace=True)
contracts.set_index('customer_id', inplace=True)
contracts
```

```python
df = contracts.join(customers, how='inner') 
df
```

```python
df['deposit'] = df['deposit'].str.replace('$', '')
df['deposit'] = df['deposit'].str.replace(',', '')

df['monthly_amt'] = df['monthly_amt'].str.replace('$', '')
df['monthly_amt'] = df['monthly_amt'].str.replace(',', '')

df
```

```python
df[["deposit", "monthly_amt"]] = df[["deposit", "monthly_amt"]].apply(pd.to_numeric)
print (df)
```

```python
df["total_monthlyamt"] = df["contract_term"] * df["monthly_amt"]
df["total_sales"] = df["total_monthlyamt"] + df["deposit"]
df
```

```python
sales_df = df.groupby(by=["state"]).sum("total_sales")
sales_df[['total_sales']]
```

```python
sales_df = sales_df.sort_values(by="state", ascending=False).head(10)
sales_df
```
